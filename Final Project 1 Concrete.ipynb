```python
import matplotlib.pyplot as plt          
import seaborn as sns
import numpy as np
import pandas as pd
print('lib import completed')
```

```python
data= pd.read_csv('Concrete_Data - copy.csv')
X = data.iloc[:,[0,1,2,3,4,5,6]].values                  #independent values
y = data.iloc[:, 7].values                               #dependent values
data
```

```python
X=X.astype(np.float64)
Y=y.astype(np.float64)
```

```python
#model evalution 
from sklearn.model_selection import train_test_split
Itrain,Itest,Dtrain,Dtest= train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression            #importing Linear Regression
LR=LinearRegression()
LR.fit(Itrain,Dtrain)
```

```python
#prediction of test data
Yprediction=LR.predict(Itest)
```

```python
import statsmodels.formula.api as sm                            #Importing ststsmodel
import statsmodels.regression.linear_model as SM
```

```python
I=np.append(arr=np.ones([1030,5]).astype(int),values=X,axis=1)               #appending input values 
print(I)
```

```python
#endog-dependent
#exdog-independent
Iopt=I[:,[0,1,2,3,4,5,6]]
Optimizer=SM.OLS(endog=Y,exog=Iopt).fit()                              #OLS = ordinary least squares
Optimizer.summary()
```

```python
#endog-dependent
#exdog-independent
Iopt=I[:,[0,1,2,3]]                           #just taking most significant values.
Optimizer=SM.OLS(endog=Y,exog=Iopt).fit()
Optimizer.summary()
```
