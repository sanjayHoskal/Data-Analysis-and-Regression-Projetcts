# Regression 
Regression is a type of supervised learning in ML that helps in mapping a predictive relationship between labels and data points. The top types of regression algorithms in ML are linear, polynomial, logistic, stepwise, etc. 

```python
# Importing Libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Housing price prediction.csv')    #importing data      
data
```

```python
# Dropping Address Column
data.drop(['Address'],axis=1,inplace=True)
```

```python
data
```

```python
#independent
# Putting feature variable to X
X = data.iloc[:,[0,1,2,3,4]].values
#Dependent
# Putting response variable to y
Y = data.iloc[:,5].values
```

```python
# Checking for Null Values
data.info()
```

```python
# Getting the summary of Data
data.describe()
```

```python
from sklearn import preprocessing                      #Importing scikit learn  
pre_process = preprocessing.StandardScaler()           
```

```python
X = pd.DataFrame(pre_process.fit_transform(X))         #fit_transform used to training data so that we can scale the training 
                                                       #data and also learn the scaling parameters of that data.    
```

```python
data.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split                   #importing test_train_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8 ,test_size = 0.2, random_state=0)
#given train size 80% & test size 20% for getting higher accuracy.
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#import linear regression
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,Y_train)
```

```python
#prediction of test data
Yprediction=LR.predict(X_test)
```

```python
import statsmodels.formula.api as sm                        #importing statsmodels & Application Program Interface.
import statsmodels.regression.linear_model as SM
```

```python
I=np.append(arr=np.ones([5000,5]).astype(int),values=X,axis=1)           #inserting values for test
print(I)
```

```python
#endog-dependent
#exdog-independent
Iopt=I[:,[0,1,2,3,4]]
Optimizer=SM.OLS(endog=Y,exog=Iopt).fit()
Optimizer.summary()
```

```python
Iopt=I[:,[2,3,4]]                                   #Randomly checking multiple times.
Optimizer=SM.OLS(endog=Y,exog=Iopt).fit()
Optimizer.summary()
```
